```python
x = 3547000
t = list(range(1, x))
y= [i for i in t if x % i == 0]

def is_prime_number(y):
    if y >= 2:
        for z in range(2,y):
            if not ( y % z ):
                return False
    else:
	    return False
    return True

new_list = [i for i in y if is_prime_number(i)]

Final_list = new_list[-2:]

print(Final_list)
```
